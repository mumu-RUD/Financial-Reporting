```python
# Importing pandas library

import pandas as pd
```

```python
# Reading excel file

df = pd.read_excel('SEC EDGAR Manual Report.xlsx')
```

```python
# Calculating the Revenue Growth and Net Income Growth in percentage
# For Revenue Growth, ((Current Year's Revenue - Previous Year's Revenue)/ Previous Year's Revenue) * 100
# For Income Growth, ((Current Year's Income - Previous Year's Income)/ Previous Year's Income) * 100

df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df
```

<br> For Apple, the revenue and income increased from 2022 to 2023 but decreased in 2024.</br>
<br> For Microsoft, there was steady decrease in revenue and income from 2022 to 2023 and 2023 to 2024.</br>
<br> For Tesla, there as a decrease in revenue and income from 2021 to 2022 but increased 2022 to 2023.</br>
<br> (In simple terms revenue is what company earns and income is the total revenue minus the total expenses).</br>

```python
# Additional Financial KPI Metrics

df['Net Profit Margin'] = df['Net Income']/df['Total Revenue'] * 100
df['Debt-to-Assets Ratio'] = df['Total Liabilities']/ df['Total Assets'] * 100
df['Return on Assets'] = df['Net Income']/ df['Total Assets'] * 100
df
```

<br> From observation, the <b> Net Profit Margin </b> is consistently high in all 3 years from Tesla which means that Tesla is good in its strong cost control. </br> 
<br> The <b> Debt-to-Assets Ratio </b> has drastically reduced over the years for Microsoft, which shows that the company is becoming less vulnerable to economic downturns. But, for Tesla, it seems to be vulnerable to economic downturns </br>
<br> Tesla seems to have a good <b> Return on Assets </b> overall which indicated its return of assets are worth. </br>
